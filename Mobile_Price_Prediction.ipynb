```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train_data = df
test_data = df1
```

```python
if 'id' in test_data.columns:
    test_data = test_data.drop(columns=['id'])
```

```python
common_columns = train_data.columns.intersection(test_data.columns)
```

```python
X_train = train_data[common_columns[:-1]]  # Exclude target from training features
y_train = train_data.iloc[:, -1]  # Target variable for training
```

```python
X_test = test_data[common_columns[:-1]]  # Test features (make sure they match with training)
y_test = test_data.iloc[:, -1]  # Target variable for testing
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
print(f'Predictions: {y_pred}')
```
